```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
plt.rcParams['text.usetex'] = False
plt.rcParams['font.family'] = ['Times New Roman']
```

```python
def plot_mean_se(y, y_lower, y_upper, color="black", alpha=1, label="", line_width=4):
    x = np.arange(1, len(y)+1)
    plt.plot(x, y, color=color, label=label, linewidth=line_width)
    plt.fill_between(x, y_lower, y_upper, alpha=alpha, color=color)
```

```python
l_episodes = 120
four_df = pd.read_csv(
    "./out/fourrooms_learning_curves.csv"
)

plt.figure(figsize=(15, 10))
plt.tick_params(labelsize=20)
plot_mean_se(
    four_df["DYN-HUMAN-mean"][:l_episodes],
    four_df["DYN-HUMAN-lower"][:l_episodes],
    four_df["DYN-HUMAN-upper"][:l_episodes], "red", 0.5,
    "DYN-HUMAN", 2
)
plot_mean_se(
    four_df["STATIC-HUMAN-mean"][:l_episodes],
    four_df["STATIC-HUMAN-lower"][:l_episodes],
    four_df["STATIC-HUMAN-upper"][:l_episodes], "orange", 0.5,
    "STATIC-HUMAN", 2
)
# plot_mean_se(
#     four_df["DYN-RANDOM-mean"][:l_episodes],
#     four_df["DYN-RANDOM-lower"][:l_episodes],
#     four_df["DYN-RANDOM-upper"][:l_episodes], "red", 0.5,
#     "DYN-RANDOM", 2
# )
plot_mean_se(
    four_df["SARSA-mean"][:l_episodes],
    four_df["SARSA-lower"][:l_episodes],
    four_df["SARSA-upper"][:l_episodes], "black", 0.5,
    "SARSA", 2
)
# plot_mean_se(
#     four_df["LINRS1-mean"][:l_episodes],
#     four_df["LINRS1-lower"][:l_episodes],
#     four_df["LINRS1-upper"][:l_episodes], "yellow", 0.5,
#     "LINRS1", 2
# )
# plot_mean_se(
#     four_df["LINRS2-mean"][:l_episodes],
#     four_df["LINRS2-lower"][:l_episodes],
#     four_df["LINRS2-upper"][:l_episodes], "orange", 0.5,
#     "LINRS2", 2
# )
# plot_mean_se(
#     four_df["LINRS3-mean"][:l_episodes],
#     four_df["LINRS3-lower"][:l_episodes],
#     four_df["LINRS3-upper"][:l_episodes], "pink", 0.5,
#     "LINRS3", 2
# )
# plot_mean_se(
#     four_df["LINRS4-mean"][:l_episodes],
#     four_df["LINRS4-lower"][:l_episodes],
#     four_df["LINRS4-upper"][:l_episodes], "green", 0.5,
#     "LINRS4", 2
# )
plot_mean_se(
    four_df["LINRS-mean"][:l_episodes],
    four_df["LINRS-lower"][:l_episodes],
    four_df["LINRS-upper"][:l_episodes], "green", 0.5,
    "LINRS", 2
)
plot_mean_se(
    four_df["NRS-mean"][:l_episodes],
    four_df["NRS-lower"][:l_episodes],
    four_df["NRS-upper"][:l_episodes], "yellow", 0.5,
    "NRS", 2
)
plt.ylabel("steps per episode", fontsize=28)
plt.xlabel("episode", fontsize=28)
plt.xlim(1, l_episodes)
plt.legend(fontsize=28)
plt.plot()
# plt.savefig("out/4rooms-learning-curves.pdf", format="pdf")
```
