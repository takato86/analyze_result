```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
plt.rcParams['text.usetex'] = False
plt.rcParams['font.family'] = ['Times New Roman']
```

```python
def plot_mean_se(y, y_lower, y_upper, color="black", alpha=1, label="", line_width=4):
    x = np.arange(1, len(y)+1)
    plt.plot(x, y, color=color, label=label, linewidth=line_width)
    plt.fill_between(x, y_lower, y_upper, alpha=alpha, color=color)
```

```python
l_episodes = 250
four_df = pd.read_csv(
    "./out/learning_curve.csv"
)

plt.figure(figsize=(15, 10))
plt.tick_params(labelsize=20)
plot_mean_se(
    four_df["DYN-HUMAN-01-mv"][:l_episodes],
    four_df["DYN-HUMAN-01-lower"][:l_episodes],
    four_df["DYN-HUMAN-01-upper"][:l_episodes], "blue", 0.5,
    "DYN-HUMAN-01", 2
)
plot_mean_se(
    four_df["DYN-HUMAN-001-mv"][:l_episodes],
    four_df["DYN-HUMAN-001-lower"][:l_episodes],
    four_df["DYN-HUMAN-001-upper"][:l_episodes], "red", 0.5,
    "DYN-HUMAN-001", 2
)
plot_mean_se(
    four_df["DYN-HUMAN-005-mv"][:l_episodes],
    four_df["DYN-HUMAN-005-lower"][:l_episodes],
    four_df["DYN-HUMAN-005-upper"][:l_episodes], "yellow", 0.5,
    "DYN-HUMAN-005", 2
)
plot_mean_se(
    four_df["DYN-HUMAN-0001-mv"][:l_episodes],
    four_df["DYN-HUMAN-0001-lower"][:l_episodes],
    four_df["DYN-HUMAN-0001-upper"][:l_episodes], "green", 0.5,
    "DYN-HUMAN-0001", 2
)
plot_mean_se(
    four_df["DYN-HUMAN-0005-mv"][:l_episodes],
    four_df["DYN-HUMAN-0005-lower"][:l_episodes],
    four_df["DYN-HUMAN-0005-upper"][:l_episodes], "orange", 0.5,
    "DYN-HUMAN-0005", 2
)
plt.ylabel("steps per episode", fontsize=28)
plt.xlabel("episode", fontsize=28)
plt.xlim(1, l_episodes)
plt.legend(fontsize=28)
plt.plot()
# plt.savefig("out/4rooms-learning-curves.pdf", format="pdf")
```
